```python
def plot_all(function, description):
    print("PFS Benefit:")
    mann_whitney_hyper_label_printer(cohort.plot_benefit(on={description: function}),
                                     label="pfs_%s" % function.__name__)
    plt.show()
    df = cohort.as_dataframe(on=function)
    print("OS Benefit:")
    mann_whitney_hyper_label_printer(cohort.plot_benefit_os(on={description: function}),
                                     label="os_%s" % function.__name__)
    plt.show()
    survival_hyper_label_printer(cohort.plot_survival(on={description: function}, how="pfs"),
                                 label="pfs_%s_curve" % function.__name__)
    plt.show()
    survival_hyper_label_printer(cohort.plot_survival(on={description: function}, how="os"),
                                 label="os_%s_curve" % function.__name__)
    plt.show()
```

```python
for function, description in [
    (missense_snv_count, "# Missense SNVs / MB"),
    (expressed_missense_snv_count, "# Expr. Missense SNVs / MB"),
    (neoantigen_count, "# Neoantigens / MB"),
    (expressed_neoantigen_count, "# Expr. Neoantigen / MB")]:
    plot_all(function, description)
```
